# **Opening a file**

It is good practice to use the `with` keyword when dealing with file objects. The advantage is that the file is properly closed after its suite finishes, even if an exception is raised at some point.

    >>> with open(filename, [access_mode]) as f:
    ...    read_data = f.read()
    
**file_name** − The file_name argument is a string value that contains the name of the file that you want to access.

**access_mode** − The access_mode determines the mode in which the file has to be opened, i.e., read, write, append, etc. This is optional parameter and the default file access mode is read (r).

Once the indentation is removed the file object will be closed.  After a file object is closed attempts to use the file object will automatically fail.

The rest of the examples in this section will assume that a file object called `f` has already been created.

## Reading from a file: `.read()`

The `read()` method of a file object reads the entire content of the file and returns it as a single string. An optional argument can be passed to `read()` to specify the number of characters or bytes to read.

```
    with open("example.txt", "r") as file:
        content = file.read()
        print(content)
```

## Reading from a file: `.readline()`

`readline()` is a method primarily used with file objects to read a single line of text from a file. It reads characters from the current position in the file up to and including the next newline character (`\n`) or the end of the file.

**Key features of readline()**:

**Reads one line at a time**:
Unlike `read()` which reads the entire file content as a single string, `readline()` focuses on retrieving content line by line.

**Includes the newline character**:
The returned string typically includes the trailing newline character (`\n`) if present in the file. This can be removed using `strip()` if desired.

**Optional `size` argument**:
You can provide an optional size argument to limit the number of bytes read from a line. If the size is less than the length of the line, only that many bytes will be returned.

**Returns an empty string at `EOF`**:
When the end of the file (`EOF`) is reached and no more lines can be read, `readline()` returns an empty string (`''`).

```
# Open a file in read mode
with open('my_file.txt', 'r') as file:
    # Read the first line
    line1 = file.readline()
    print(f"First line: {line1.strip()}") # Using strip() to remove newline

    # Read the second line
    line2 = file.readline()
    print(f"Second line: {line2.strip()}")

    # Loop through remaining lines
    for line in file: # More common and efficient way to iterate lines
        print(f"Remaining line: {line.strip()}")
```

# **Writing to a file**

The `open()` function is used to open a file. It takes at least two arguments: the file path and the mode. Common modes for writing include:

**`"w"` (write mode)**:
Opens the file for writing. If the file exists, its content is truncated (emptied) before writing. If the file does not exist, a new file is created.

**`"a"` (append mode)**:
Opens the file for appending. Data is written at the end of the file. If the file does not exist, a new file is created.

```
# Example: Opening a file in write mode
with open("my_file.txt", "w") as file:
    # Operations to write to the file
    pass

# Example: Opening a file in append mode
with open("log.txt", "a") as file:
    # Operations to append to the file
    pass
```

**Writing Content to the File**:

**`file.write(string)`**: Writes a single string to the file. This method does not automatically add a newline character, so you need to include `\n` if you want to write multiple lines.

```
with open("my_file.txt", "w") as file:
    file.write("This is the first line.\n")
    file.write("This is the second line.")
```
**`file.writelines(iterable_of_strings)`**: Writes a list or other iterable of strings to the file. Like `write()`, it does not automatically add newline characters, so you must include them in your strings if needed.

```
lines = ["Line 1\n", "Line 2\n", "Line 3"]
with open("another_file.txt", "w") as file:
    file.writelines(lines)
```

## Create a Time Series Plot

Using the data contained in the file **time_and_temp.txt**
  

1.   plot the data in the file
2.   use the text information in the file to construct a title, x-axis label and y-axis label
3.   construct a plot code to draw a dashed line with `"o"` as data markers and the color set to green.  

Notice the plot shows a time history of the data that was collected.

````python
# Time Series Plot
from matplotlib import pyplot as plt
import numpy as np

# Read the time_and_temp.txt file and extract the time and temperature data into two separate lists
with open('time_and_temp.txt', 'r') as file:
    temp_data = file.readlines()
    
temp_times = []
temps = []

for line in temp_data[1:]:
    time,temp = line.split()
    
    temp_times.append(float(time))
    temps.append(float(temp))

# Setting the plots size
plt.figure(figsize = (20, 5))

# Plotting the time vs temperature data
plt.plot(temp_times, temps, color = 'green', marker = 'o', linestyle = '--')

plt.title('Temperature vs Time')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')


plt.show()
````

## Create a Scatter Plot

Using the data in **temp_and_pressure.txt** plot the data in the file and use the text information in the file to construct a title, x-axis label and y-axis label.

1.   Construct a plotting code with no lines drawn and `"o"` as data markers and the color set to red.

2.   Locate the centroid of the plot by computing the average value of the temperature data and pressure data. The plot centroid has coordinates (average temperature, average pressure) .

3.   Label this point with the text ‘Data Centroid’.

4.   Finally plot horizontal and vertical lines through the centroid.


````python
# Scatter Plot
from matplotlib import pyplot as plt
import numpy as np

# Read the temp_and_pressure.txt file and extract the time and pressure data into two separate lists
with open('temp_and_pressure.txt', 'r') as file:
    pressure_data = file.readlines()
    
pressure_temps = []
pressures = []

for line in pressure_data[1:]:
    time,pressure = line.split()
    
    pressure_temps.append(float(time))
    pressures.append(float(pressure))

# Find the centroid of the data points
centroid_y = np.mean(pressures)
centroid_x = np.mean(pressure_temps)
    
# Setting size for the plot
plt.figure(figsize = (20, 8))

# Creating a scatter plot
plt.scatter(pressure_temps, pressures, color = 'red', marker = 'o')

# Plotting the centroid
plt.scatter(centroid_x, centroid_y, color = 'blue', marker = 'x', s = 100, label = 'Data Centroid')

# Drawing a horizontal and vertical line through the centroid
plt.axhline(y = centroid_y, color = 'blue', linestyle = '--', linewidth = 1)
plt.axvline(x = centroid_x, color = 'blue', linestyle = '--', linewidth = 1)


plt.legend()

plt.title('Pressure vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Pressure (atm)')

plt.show()
````

## Subplots.

Use data in the **time_temp_pressure_volume.txt** file to draw a 3x1 (3 rows 1 column) subplot.  

The first subplot should contain temperature

*   The first subplot should contain temperature versus time,
*   The second subplot should contain pressure versus time
*   The third subplot should contain volume versus time.
*   use the data in the file to construct a title, x-axis, and y-axis label for each plot.
*   For each subplot draw a horizontal line through the average value of the data for the subplot.



````python
# Subplots

# Read the time_temp_pressure_volume.txt file and extract the time, temperature, pressure, and volume data into four separate lists
with open('time_temp_pressure_volume.txt', 'r') as file:
    all_data = file.readlines()
    
all_times = []
all_temps = []
all_pressures = []
all_volumes = []

for line in all_data[1:]:
    time, temp, pressure, volume = line.split()
    
    all_times.append(float(time))
    all_temps.append(float(temp))
    all_pressures.append(float(pressure))
    all_volumes.append(float(volume))
    
# Finding the average value of each variable to use for the horizontal lines in the subplots
avg_temp = np.mean(all_temps)
avg_pressure = np.mean(all_pressures)
avg_volume = np.mean(all_volumes)

# Create subplots

# Setting size for the subplots
plt.figure(figsize = (20, 10))

# Subplot 1 - Temperature vs Time
plt.subplot(311)
plt.plot(all_times, all_temps, color = 'green', marker = 'o', linestyle = '--')
plt.title('Temperature vs Time')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')
plt.axhline(y = avg_temp, color = 'blue', linestyle = '--', linewidth = 1, label = 'Average Temperature')
plt.legend()

# Subplot 2 - Pressure vs Time
plt.subplot(312)
plt.plot(all_times, all_pressures, color = 'red', marker = 'o', linestyle = '--')
plt.title('Pressure vs Time')
plt.xlabel('Time (hours)')
plt.ylabel('Pressure (atm)')
plt.axhline(y = avg_pressure, color = 'purple', linestyle = "--", linewidth = 1, label = 'Average Pressure')
plt.legend()

# Subplot 3 - Volume vs Time
plt.subplot(313)
plt.plot(all_times, all_volumes, color = 'blue', marker = 'o', linestyle = '--')
plt.title('Volume vs Time')
plt.xlabel('Time (hours)')
plt.ylabel('Volume (L)')
plt.axhline(y = avg_volume, color = 'red', linestyle = "--", linewidth = 1, label = 'Average Volume')
plt.legend()

# Show the subplots
plt.tight_layout()
plt.show()
````

## Generate a sales report using pie charts
 Your boss wants you to create a monthly sales report analyzing the performance of the sales staff.

* The sales persons names are Bill, Sue, Mary, Jane, Dave, and John.
* Using the data in the file **sales_data.txt** create 2 pie chart plots in one plotting window using the subplot command to create a window with one row and two columns.
* Subplot number 1 should contain a pie chart that shows the total sales for the month for each salesperson.
* Subplot number 2 should contain a pie chart that shows the total dollar amount for the month for each salesperson.
* For each pie chart “explode” the slice with the maximum value.


````python
# Pie Chart

with open('sales_data.txt', 'r') as file:
    sales_data = file.readlines()

sales = {}
dollar_amount = {}


for line in sales_data[1:]:
    salesperson, dollars = line.split()
    
    # Call the key of the first dictionary "sales" and the key of the second dictionary "dollar_amount" 
    # If the key already exists, increase its value by 1 for sales or add the dollar amount to the existing value for dollar_amount. 
    # If the key does not exist, initialize it with 1 for sales and the dollar amount for dollar_amount.
    sales.update({salesperson: sales.get(salesperson, 0) + 1})
    dollar_amount.update({salesperson: dollar_amount.get(salesperson, 0) + float(dollars)})
    
total_sales = sum(sales.values())
total_dollar_amount = sum(dollar_amount.values())
# Sizing the plots
plt.figure(figsize = (15, 7))

# Subplot 1 - Number of Sales per Salesperson
plt.subplot(121)
plt.pie(sales.values(), labels = sales.keys(), startangle = 140, autopct=lambda p: f'{p:.1f}%\n({int(p * total_sales / 100)})')

plt.title('Number of Sales per Salesperson')

# Subplot 2 - Total Dollar Amount per Salesperson
plt.subplot(122)
plt.pie(dollar_amount.values(), labels = dollar_amount.keys(), startangle = 140, autopct = lambda p: f'{p:.1f}%\n({int(p * total_dollar_amount / 100)})')

plt.title('Total Dollar Amount per Salesperson')
````

## Portland Weather Data

Using the portlandweather.txt file:

1.   Look at the data and determine what Python data structures you'll use to store it. We will only use the data from the 'DATE', 'PRCP', 'SNOW', 'TMAX', and 'TMIN' columns
2.   "Clean" the data:
     * Change the format for the date, if desired
     * PRCP and SNOW are given in mm's, change to inches
     * TMAX and TMIN have had one decimal place removed and are then in Celsius, change to Farhenheit
3.  Plot Data using Subplots
     * In plot 1 create a scatterplot of high temps for the entire year.  In plot 2 do the same but for low temps.
     * Create a subplot with histograms of precip and snow for April
     * Create a horizontal bar graph with both average high and average low temps for each month, on the same plot.




````python
# Portland Weather

dates = []
prcp = []
snow = []
tmax = []
tmin = []

days_of_year = [i for i in range(1, 366)]
with open('PortlandWeather2013.txt', 'r') as file:
    weather_data = file.readlines()

for line in weather_data[2:]:
    station, date, pr, snwd, sn, tx, tn = line.split()
    
    # turn strings into floats or integers as needed
    pr = int(pr)
    sn = int(sn)

    new_date = date[4:6] + '/' + date[6:8] + '/' + date[0:4]  # convert from YYYYMMDD to MM/DD/YYYY
    
    new_pr = int(pr)/25.4  # convert from mm to inches
    new_sn = int(sn)/25.4  # convert from mm to inches
    
    new_pr = format(new_pr, '.2f')
    new_sn = format(new_sn, '.2f')
    
    # correcting the decimal point placement for tmax and tmin and turning them into floats
    tx = tx[:len(tx)-1] + '.' + tx[len(tx)-1:]
    tn = tn[:len(tn)-1] + '.' + tn[len(tn)-1:]  
    tx = float(tx)
    tn = float(tn)
    
    # celsius to fahrenheit conversion
    new_tx = (tx * 9/5) + 32
    new_tn = (tn * 9/5) + 32
    
    dates.append(new_date)
    prcp.append(float(new_pr))
    snow.append(float(new_sn))
    tmax.append(float(new_tx))
    tmin.append(float(new_tn))

# Turning weather data into dictionaries
weather_dict = {'Date': dates, 'Precipitation (inches)': prcp, 'Snowfall (inches)': snow, 'Max Temperature (°F)': tmax, 'Min Temperature (°F)': tmin}

# Setting size for the plots
plt.figure(figsize = (20, 10))

# # Subplot 1 - Scatterplots 1
plt.subplot(211)
plt.scatter(days_of_year, tmax, color = 'blue', marker = 'o')
plt.xlabel('Date')
plt.ylabel('Max Temperature (°F)')
plt.title('Daily Max Temperature in Portland, 2013')

# Subplot 1 - Scatterplots 2
plt.subplot(212)
plt.scatter(days_of_year, tmin, color = 'red', marker = 'o')
plt.xlabel('Date')
plt.ylabel('Min Temperature (°F)')
plt.title('Daily Min Temperature in Portland, 2013')

plt.tight_layout()

````

````python
# Plot 3 - Bar graph 1 

# Setting size for the plots
plt.figure(figsize = (20, 10))

# Sorting averages temps for each month, high and low 
avg_monthly_temps_high = []
avg_monthly_temps_low = []
    
# Define the number of days in each month for a non-leap year
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

start = 0

for days in days_in_month:
    end = start + days
    avg_high = np.mean(tmax[start:end])
    avg_low = np.mean(tmin[start:end])
    avg_monthly_temps_high.append(round(avg_high, 2))
    avg_monthly_temps_low.append(round(avg_low, 2))
    start = end

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

y_indexes = np.arange(len(months))  

width = 0.35  

plt.barh(y_indexes, avg_monthly_temps_high, color = 'blue', height = width, label = 'Avg High Temp')
plt.barh(y_indexes + width, avg_monthly_temps_low, color = 'red', height = width, label = 'Avg Low Temp')
plt.yticks(ticks = y_indexes, labels = months)    
    
plt.xlabel('Average Temperature (°F)')
plt.title('Average Monthly High and Low Temperatures in Portland, 2013')
plt.legend()
plt.tight_layout()
````
