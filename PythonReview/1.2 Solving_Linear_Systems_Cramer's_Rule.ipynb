# Determinants

In mathematics, the `determinant` is a `scalar` value that is a function of the entries of a square matrix. It allows characterizing some properties of the matrix and the linear map represented by the matrix. The determinant of a product of matrices is the product of their determinants (the preceding property is a corollary of this one). The determinant of a matrix A is denoted det(A), det A, or |A|.

Determinants occur throughout mathematics. For example, a matrix is often used to represent the coefficients in a system of linear equations, and determinants can be used to solve these equations (Cramer's rule)

Create a function that returns the determinant of a 2 x 2 matrix.

```python
# create a determinant function that takes a 2x2 matrix (ie. list) as an
# argument and returns the determinant of the given matrix

matrix = [[1, 2], [3, 4]]

def determinant(matrix):
    determinant = (matrix[0][0] * matrix[1][1]) - (matrix[0][1] * matrix[1][0])
    return determinant

print(determinant(matrix))
```

## Cramer's Rule

In linear algebra, Cramer's rule is an explicit formula for the solution of a system of linear equations with as many equations as unknowns, valid whenever the system has a unique solution. It expresses the solution in terms of the determinants of the (square) coefficient matrix and of matrices obtained from it by replacing one column by the column vector of right-sides of the equations.

Then, with help of determinants, x and y can be found with Cramer's rule as:

### To do:

Create a function that returns a tuple containing the solutions to a system of 2 equations with 2 unknowns in Standard Form:

    a1x + b1y = c1
    a2x + b2y = c2

    5x + 5y = -20
    4x - 4y = 8

Has a solution of (-1, -3)

```python
# create a determinant function that takes a 2x3 matrix (ie. list) as an
# argument and returns a tuple containing the values of the variables

matrix = [[5, 5, -20], [4, -4, 8]]

def cramer(matrix):
    a1 = matrix[0][0]
    b1 = matrix[0][1]
    c1 = matrix[0][2]
    a2 = matrix[1][0]
    b2 = matrix[1][1]
    c2 = matrix[1][2]
    
    determinant = (a1 * b2) - (a2 * b1)
    determinant_x = (c1 * b2) - (c2 * b1)
    determinant_y = (a1 * c2) - (a2 * c1)
    
    x = determinant_x / determinant
    y = determinant_y / determinant
    
    print(determinant, determinant_x, determinant_y)
    
    return (x, y)

print(cramer(matrix))
    
```
