```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import data_analytics_lib as dal
```

# Open, read data from the text file, and store it in lists

```python
with open('website_stats.txt', 'r') as file:
    data = file.readlines()
    
web_a_times = []
web_a_clicks = []
web_b_times = []
web_b_clicks = []

# Convert data to a list of floats
for line in data[1:]:  # Skip header line
    weba_time, weba_clicks, webb_time, webb_clicks = line.split()
    
    web_a_times.append(float(weba_time))
    web_a_clicks.append(float(weba_clicks))
    web_b_times.append(float(webb_time))
    web_b_clicks.append(float(webb_clicks))
    
```

# Website A - Time Spent Data

```python
# Use my imported data analytics library to analyze the data

# Calculate and print summary statistics for Web A time spent

web_a_time_summary = dal.main_stats(web_a_times)

# Making dataframe 
web_a_time_df = pd.DataFrame(web_a_time_summary, index=[0])
web_a_time_df
```

```python
time_spent_a_plt = dal.regular_plot(web_a_times, title="Website A - Time Spent Regular Plot", xaxis_label="User", yaxis_label="Time Spent (mins)", color="blue")
```

```python
time_spent_a_histogram = dal.histogram(web_a_times, title="Website A - Time Spent Histogram", xaxis_label="Time Spent (mins)", yaxis_label="User", color ="blue", edgecolor="black")
```

```python
time_spent_a_boxplot = dal.boxplot(web_a_times, title="Website A - Time Spent Boxplot", xaxis_label="Time A", yaxis_label="Time Spent (mins)")
```

# Website A - Average Clicks Per Minute

```python
average_clicks_per_minute_a_list = []

for i in range(len(web_a_times)):
    if web_a_times[i] > 0:
        avg_clicks_per_minute_a = web_a_clicks[i] / web_a_times[i]
    else:
        avg_clicks_per_minute_a = 0  # Avoid division by zero
    
    avg_clicks_per_minute_a = round(avg_clicks_per_minute_a, 2)
    average_clicks_per_minute_a_list.append(avg_clicks_per_minute_a)
    
average_clicks_per_minute_a_summary = dal.main_stats(average_clicks_per_minute_a_list)
average_clicks_per_minute_a_df = pd.DataFrame(average_clicks_per_minute_a_summary, index=[0])
average_clicks_per_minute_a_df
```

```python
avg_quartiles_a_summary = dal.quartiles(average_clicks_per_minute_a_list)
avg_quartiles_a_df = pd.DataFrame(avg_quartiles_a_summary, index=[0])
avg_quartiles_a_df
```
