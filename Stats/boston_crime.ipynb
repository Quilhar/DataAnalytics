```python
import numpy as np
import matplotlib.pyplot as plt
import data_analytics_lib as dal
import random
import pandas as pd
```

```python
def read_data(filepath):
    
    data = {}

    with open(filepath, "r", encoding='cp1252') as file:
        header = file.readline().strip().split(",")
        
        for key in header:
            data[key] = []
            
        for line in file:
            
            lines = line.strip().split(",")
            for i, key in enumerate(header):
                data[key].append(lines[i])

    return data

main_crime_data = read_data("/Users/244213/Desktop/DataAnalytics/Stats/crime.csv")

print(main_crime_data['YEAR'][:23])


main_crime_data_df = pd.DataFrame(main_crime_data)

# main_crime_data_df.head(10)
```

```python
def get_crime_data_by_specific_year(data, year):
    crime_data = {
        year: {
            "MONTH": [],
            "DAY_OF_WEEK": []
        }
    }

    for i in range(len(data["YEAR"])):
        if data["YEAR"][i] == year:
            crime_data[year]["MONTH"].append(data["MONTH"])
            crime_data[year]["DAY_OF_WEEK"].append(data["DAY_OF_WEEK"])

    return crime_data

crime_data_2016 = get_crime_data_by_specific_year(main_crime_data, 2016)

print(crime_data_2016)
```
